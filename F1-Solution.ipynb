```python
import numpy as np
import matplotlib.pyplot as plt
```

(a) sol: 
The sine-cosine Fourier expandsion:
\begin{equation}
f(x) = \frac{1}{2}a_{0} + \sum_{n=1}^{\infty}a_{n}\cos{\frac{2n\pi x}{L}} +\sum_{n=1}^{\infty}b_{n}\sin{\frac{2n\pi x}{L}}
\end{equation}
\
In this case, $L = 2\pi$:
\begin{equation}
f(x) = \frac{1}{2}a_{0} + \sum_{n=1}^{\infty}a_{n}\cos{nx} +\sum_{n=1}^{\infty}b_{n}\sin{nx}
\end{equation}
\
\begin{equation}
a_{0} = \frac{1}{\pi}\int^{\pi}_{-\pi} f(x) dx = \frac{1}{\pi}\int^{\pi}_{0} x dx = \frac{\pi}{2}
\end{equation}
\
\begin{eqnarray}
a_{n} &=& \frac{2}{L}\int^{L/2}_{-L/2} f(x)\cos \frac{2n\pi}{L} dx\\
&=& \frac{1}{\pi}\int^{\pi}_{0} x\cos(nx)dx\\
&=& \frac{1}{\pi}(\frac{1}{n}x\sin(x) \Big{|}^{\pi}_{0}- \int^{\pi}_{0}\frac{1}{n}\sin(nx)dx)\\
&=& \frac{1}{\pi}\frac{1}{n^{2}}(\cos{n\pi}-1)\\
&=& \frac{(-1)^{n}-1}{\pi n^{2}}
\end{eqnarray}
\
\begin{eqnarray}
b_{n} &=& \frac{2}{L}\int^{L/2}_{-L/2} f(x)\sin \frac{2n\pi}{L} dx\\
&=& \frac{1}{\pi}\int^{\pi}_{0} x\sin(nx)dx\\
&=& \frac{1}{\pi}(-\frac{1}{n}x\cos(x) \Big{|}^{\pi}_{0}- \int^{\pi}_{0}\frac{1}{n}\cos(nx)dx)\\
&=& -\frac{1}{\pi}\frac{\pi}{n}\cos{n\pi}\\
&=& \frac{(-1)^{n+1}}{n}
\end{eqnarray}
We have:
\begin{equation}
f(x) = \frac{\pi}{4} + \sum_{n=1}^{\infty}\frac{(-1)^{n}-1}{\pi n^{2}}\cos{nx} +\sum_{n=1}^{\infty}\frac{(-1)^{n+1}}{n}\sin{nx}
\end{equation}

```python
pi = np.pi
xs = np.linspace(-pi,pi,1000)
fig = plt.figure(figsize=(12,6))

###Fourier-series###
plt.subplot(1,2,1)
steps = [5,10,30]
fs = []
for step in steps:
    fs = []
    for x in xs:
        f=pi/4
        for n in range(1,step+1):
            an = ((-1)**n -1)/(pi*(n**2))
            bn = ((-1)**(n+1))/n
            f = f + an*np.cos(n*x) + bn*np.sin(n*x)
        fs.append(f)
    plt.plot(xs,fs,label = 'n= %d'%step)
    plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fourier expandsion')
####################

###exact###
plt.subplot(1,2,1)
temp1 = [0 for x in xs if x<0]
temp2 = [x for x in xs if x>0] 
ys = temp1+temp2
plt.plot(xs,ys,label = 'exact')
plt.legend()
plt.subplot(1,2,2)
plt.plot(xs,ys,label = 'exact')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exact')
##########
```

(c) sol:\
By the Dirichlet's theorem, we know f(x=0) converge to zero, we have:
\begin{equation}
f(0) = \frac{\pi}{4} + \sum_{n=1}^{\infty}\frac{(-1)^{n}-1}{\pi n^{2}} = 0
\end{equation}
The later term vanishes when n is a even number.
\begin{eqnarray}
\frac{\pi}{4} + \sum_{n=odd}^{\infty}\frac{-2}{\pi n^{2}} &=& 0\\
\sum_{n=odd}^{\infty}\frac{1}{n^{2}} &=& \frac{\pi ^2}{8}
\end{eqnarray}
